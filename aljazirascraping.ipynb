```python
%%writefile aljaziranews.py
import scrapy
class Aljazira(scrapy.Spider):
    name = "aljaziranews"
    allowed_domains = ["www.aljazeera.net"]
    start_urls = ["https://www.aljazeera.net/news/"]

    def parse(self, response):
      news=response.css('div')
      print(len(news))
      for i in news:
        yield{
          'title' :i.css('div h3 a span::text').get(),
          'description' :i.css('div div div p::text').get(),
          'date' : i.css('footer div div div div span::text').get(),
          'image' : i.css('div div img').get(),
        }
```

```python
!scrapy crawl aljaziranews -o news_data.json
```

```python
import json
with open('news_data.json','r') as file:
  data=json.load(file)
  print(len(data))
```
